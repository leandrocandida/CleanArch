```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.preprocessing import MinMaxScaler
```

```python
ticker = "^HSI"
df = yf.download(ticker,period='10y',interval='1D')
df = df.xs((ticker),level="Ticker",axis=1)
```

```python
df['Return'] = df['Close'].pct_change()
v = 5
df['HistVol'] = df['Return'].rolling(window=v).std()*100
df['FutureVolatility'] = df['HistVol'].shift(-v)

```

```python
for lag in range(1,v+1):
  df[f'lag_{lag}'] = df['HistVol'].shift(lag)
```

```python
# Calcular a Média Móvel Exponencial (MME)
df['MME'] = df['Close'].ewm(span=22, adjust=False).mean()

 # Calcular o RSL (Relative Strength Levy)
df['RSL'] = df['Close'] / df['MME']
```

```python
df['volume_medio'] = df['Volume'].rolling(window=20).mean()
```

```python
df['volume_medio_norm'] = (df['volume_medio'] - df['volume_medio'].min()) / (df['volume_medio'].max() - df['volume_medio'].min())
```

```python
#df = df.dropna().copy()
```

```python
features = [f'lag_{lag}' for lag in range(1,v+1)]
target = 'FutureVolatility'
```

```python
features.append('RSL')
features.append('volume_medio_norm')
```

```python
import numpy as np

# Corpo e range do candle
df['candle_body'] = (df['Close'] - df['Open']).abs()
df['candle_range'] = df['High'] - df['Low']

# Sombra superior e inferior
df['upper_shadow'] = df['High'] - df[['Open', 'Close']].max(axis=1)
df['lower_shadow'] = df[['Open', 'Close']].min(axis=1) - df['Low']

# Proporções (com proteção contra divisão por zero)
df['body_ratio'] = np.where(df['candle_range'] == 0, 0, df['candle_body'] / df['candle_range'])
df['upper_shadow_ratio'] = np.where(df['candle_range'] == 0, 0, df['upper_shadow'] / df['candle_range'])
df['lower_shadow_ratio'] = np.where(df['candle_range'] == 0, 0, df['lower_shadow'] / df['candle_range'])
```

```python
features.append('body_ratio')
features.append('upper_shadow_ratio')
features.append('lower_shadow_ratio')
```

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Baixar dados do Fundamentus
def get_fundamentus_data():
    url = 'https://www.fundamentus.com.br/resultado.php'
    headers = {'User-Agent': 'Mozilla/5.0'}
    r = requests.get(url, headers=headers)
    df = pd.read_html(r.text, decimal=",", thousands=".")[0]
    df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]
    return df

def convert_columns(df):
    # Corrigir colunas percentuais (todas com vírgula e %)
    percent_cols = ['roic', 'roe', 'mrg_ebit', 'cresc._rec.5a']

    for col in percent_cols:
        if col in df.columns:
            df[col] = (
                df[col].astype(str)
                .str.replace('%', '', regex=False)
                .str.replace(',', '.', regex=False)
            )
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Corrigir as demais colunas numéricas — exceto 'papel' e 'cotação'
    skip_cols = ['papel', 'cotação'] + percent_cols
    for col in df.columns:
        if col not in skip_cols:
            df[col] = (
                df[col].astype(str)
                .str.replace('%', '', regex=False)
                .str.replace('.', '', regex=False)  # remove milhar
                .str.replace(',', '.', regex=False)  # corrige decimal
            )
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Agora converte 'cotação' de forma leve, só para garantir que é float
    df['cotação'] = pd.to_numeric(df['cotação'], errors='coerce')

    return df



def normalize(series):
    min_val = series.min()
    max_val = series.max()
    if pd.isna(min_val) or pd.isna(max_val) or min_val == max_val:
        return pd.Series([0.0] * len(series), index=series.index)  # evita NaN
    return (series - min_val) / (max_val - min_val)

def invert_normalize(series):
    return 1 - normalize(series)


# Score
def calculate_score(df):
    df['score'] = (
        normalize(df['cresc._rec.5a']) * 0.4 +
        normalize(df['roic']) * 0.3 +
        normalize(df['mrg_ebit']) * 0.2 +
        invert_normalize(df['ev/ebit']) * 0.1
    )
    return df

# Estratégia Small Caps Growth
def run_small_caps_growth():
    df = get_fundamentus_data()
    df = convert_columns(df)

    # Estimar valor de mercado
    df['valor_mercado'] = (df['cotação'] * df['patrim._líq']) / df['p/vp']

    # Filtros
    df = df[
        (df['valor_mercado'] < 5_000_000_000) &  # < 5bi
        (df['liq.2meses'] > 500_000) &
        (df['roe'] > 0) &
        (df['roic'] > 0) &
        (df['p/l'] > 0) &
        (df['ev/ebit'] > 0) &
        (df['cresc._rec.5a'] > 5)
    ]

    # Score
    df = calculate_score(df)

    # Top 10
    top10 = df.sort_values('score', ascending=False).head(10)
    return top10[['papel', 'score', 'valor_mercado', 'cotação', 'roe', 'roic', 'ev/ebit', 'mrg_ebit', 'cresc._rec.5a']]

import numpy as np

def balancear_por_score_inteiro(df_ranked, capital_total=1000.0):
    ativos = df_ranked.copy()

    # Normaliza scores
    ativos['peso'] = ativos['score'] / ativos['score'].sum()

    # Capital alocado por ativo
    ativos['capital_alocado'] = ativos['peso'] * capital_total

    # Quantidade inteira (arredondando pra baixo)
    ativos['quantidade'] = np.floor(ativos['capital_alocado'] / ativos['cotação']).astype(int)

    # Total efetivamente investido por ativo
    ativos['total'] = ativos['quantidade'] * ativos['cotação']

    # Cálculo final
    total_usado = ativos['total'].sum()
    saldo_restante = capital_total - total_usado

    return ativos[['papel', 'cotação', 'score', 'quantidade', 'total']], total_usado, saldo_restante

import yfinance as yf
import pandas as pd
import time


def calcular_retorno_carteira(carteira_df, start_date="2024-01-01"):
    """
    Retorna o valor da carteira no tempo com base no preço ajustado (Adj Close).
    Valida tickers e evita que o código quebre com iloc[0].
    """
    # Garante que os tickers estejam com o sufixo .SA (para ações BR)
    carteira_df = carteira_df.copy()
    carteira_df['papel'] = carteira_df['papel'].apply(lambda x: x if x.endswith('.SA') else f"{x}.SA")

    tickers = carteira_df['papel'].tolist()
    quantidade_dict = dict(zip(carteira_df['papel'], carteira_df['quantidade']))

    df_precos = pd.DataFrame()

    for ticker in tickers:
        try:
            dados = yf.download(ticker, start=start_date, progress=False)['Close']
            if not dados.empty:
                df_precos[ticker] = dados
            else:
                print(f"[!] Sem dados para {ticker}")
        except Exception as e:
            print(f"[X] Erro ao baixar {ticker}: {e}")
        time.sleep(1)  # evita bloqueios da API

    if df_precos.empty:
        raise ValueError("Nenhum dado de preço foi carregado. Verifique os tickers e a data.")

    # Calcula o valor da carteira
    for papel in tickers:
        if papel in df_precos.columns:
            df_precos[papel] = df_precos[papel] * quantidade_dict.get(papel, 0)

    df_precos['carteira_total'] = df_precos.sum(axis=1)

    # Verifica novamente se há dados válidos
    if df_precos['carteira_total'].isna().all():
        raise ValueError("Todos os valores de carteira estão NaN. Dados incompletos ou corrompidos.")

    valor_inicial = df_precos['carteira_total'].dropna().iloc[0]
    df_precos['retorno_%'] = (df_precos['carteira_total'] - valor_inicial) / valor_inicial * 100

    return df_precos[['carteira_total', 'retorno_%']]
def calcular_retorno_carteira_vs_ibov(carteira_df, start_date="2024-01-01"):
    tickers = carteira_df['papel'].tolist()
    quantidade_dict = dict(zip(carteira_df['papel'], carteira_df['quantidade']))

    # Baixa os dados da carteira e do IBOV
    df_precos = yf.download(tickers + ['^BVSP'], start=start_date)['Close']

    # Se só um papel, converte para DataFrame
    if isinstance(df_precos, pd.Series):
        df_precos = df_precos.to_frame()

    # Renomeia as colunas
    df_precos.columns = [col.split('.')[0] for col in df_precos.columns]

    # Cálculo do valor da carteira
    for papel in tickers:
        df_precos[papel] = df_precos[papel] * quantidade_dict[papel]

    df_precos['carteira_total'] = df_precos[tickers].sum(axis=1)
    valor_inicial_carteira = df_precos['carteira_total'].iloc[0]
    df_precos['retorno_carteira_%'] = (df_precos['carteira_total'] - valor_inicial_carteira) / valor_inicial_carteira * 100

    # Cálculo do retorno do Ibovespa
    preco_ibov_inicial = df_precos['^BVSP'].iloc[0]
    df_precos['retorno_ibov_%'] = (df_precos['^BVSP'] - preco_ibov_inicial) / preco_ibov_inicial * 100

    return df_precos[['retorno_carteira_%', 'retorno_ibov_%']]

from pandas.api.types import is_datetime64_any_dtype

def calcular_retorno_com_dividendos(carteira_df, start_date="2024-01-01"):
    """
    Calcula o retorno da carteira considerando preço ajustado + dividendos reais pagos.
    Compara com o Ibovespa.
    """
    carteira_df = carteira_df.copy()
    carteira_df['papel'] = carteira_df['papel'].apply(lambda x: x if x.endswith('.SA') else f"{x}.SA")
    tickers = carteira_df['papel'].tolist()
    print(tickers)
    quantidade_dict = dict(zip(carteira_df['papel'], carteira_df['quantidade']))

    df_precos = pd.DataFrame()
    for ticker in tickers:
        try:
            dados = yf.download(ticker, start=start_date)['Close']
            if not dados.empty:
                df_precos[ticker] = dados
            else:
                print(f"[!] Sem dados para {ticker}")
        except Exception as e:
            print(f"[X] Erro ao baixar {ticker}: {e}")
        time.sleep(1)

    if df_precos.empty:
        raise ValueError("Nenhum dado de preço foi carregado. Verifique os tickers e a data.")

    # Calcula valor da carteira por dia
    for papel in tickers:
        if papel in df_precos.columns:
            df_precos[papel] = df_precos[papel] * quantidade_dict.get(papel, 0)

    df_precos['carteira_valor'] = df_precos.sum(axis=1)

    # Dividendos acumulados
    df_dividendos = pd.DataFrame(index=df_precos.index)
    df_dividendos['div_total'] = 0.0

    for papel in tickers:
        qtd = quantidade_dict[papel]
        try:
            divs = yf.Ticker(papel).dividends
            if not divs.empty and is_datetime64_any_dtype(divs.index):
                divs = divs[divs.index >= pd.to_datetime(start_date)]
            else:
                divs = pd.Series(dtype='float64')
        except Exception as e:
            print(f"[!] Erro ao pegar dividendos de {papel}: {e}")
            divs = pd.Series(dtype='float64')

        for date, valor_por_acao in divs.items():
            if date in df_dividendos.index:
                df_dividendos.loc[date, 'div_total'] += valor_por_acao * qtd
            else:
                # Se data não bate, insere com NaN e depois preenche (raro)
                df_dividendos.loc[date] = {'div_total': valor_por_acao * qtd}

    df_dividendos = df_dividendos.sort_index().reindex(df_precos.index).fillna(0)
    df_dividendos['div_acumulado'] = df_dividendos['div_total'].cumsum()

    # Valor total = preço ajustado + dividendos acumulados
    df_precos['carteira_total_com_div'] = df_precos['carteira_valor'] + df_dividendos['div_acumulado']

    # Retorno carteira com dividendos
    valor_inicial = df_precos['carteira_total_com_div'].dropna().iloc[0]
    df_precos['retorno_total_%'] = (df_precos['carteira_total_com_div'] - valor_inicial) / valor_inicial * 100

    # Baixa dados do Ibovespa
    try:
        ibov_data = yf.download('^BVSP', start=start_date)['Adj Close']
        if ibov_data.empty:
            print("[!] Dados do IBOV não disponíveis.")
            df_precos['retorno_ibov_%'] = None
        else:
            preco_ibov_inicial = ibov_data.dropna().iloc[0]
            retorno_ibov = (ibov_data - preco_ibov_inicial) / preco_ibov_inicial * 100
            df_precos = df_precos.join(retorno_ibov.rename('retorno_ibov_%'), how='left')
    except Exception as e:
        print(f"[X] Erro ao baixar IBOV: {e}")
        df_precos['retorno_ibov_%'] = None

    return df_precos[['carteira_total_com_div', 'retorno_total_%', 'retorno_ibov_%']]


def plotar_retorno(df_retorno):
    plt.figure(figsize=(12, 6))
    plt.plot(df_retorno.index, df_retorno['retorno_carteira_%'], label='Carteira Small Caps Growth', linewidth=2)
    plt.plot(df_retorno.index, df_retorno['retorno_ibov_%'], label='Ibovespa', linestyle='--', linewidth=2)
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.title('Retorno da Carteira vs Ibovespa')
    plt.xlabel('Data')
    plt.ylabel('Retorno (%)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Rodar
if __name__ == "__main__":
    resultado = run_small_caps_growth()
    display(resultado)

    df_balanced, usado, sobra = balancear_por_score_inteiro(resultado, 1000)


    print(df_balanced)
    print(f"\nTotal investido: R$ {usado:.2f}")
    print(f"Saldo restante: R$ {sobra:.2f}")

    resultado_carteira = calcular_retorno_carteira(df_balanced, start_date="2024-12-01")
    print(resultado_carteira.tail())

    df_retorno = calcular_retorno_carteira_vs_ibov(df_balanced, start_date="2024-12-01")
    plotar_retorno(df_retorno)

    df_retorno = calcular_retorno_com_dividendos(df_balanced, start_date="2024-12-01")
    plotar_retorno(df_retorno[['retorno_total_%', 'retorno_ibov_%']])


```

```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# === CONFIGURAÇÃO ===
assets = ['PETR4.SA', 'VALE3.SA', 'ITUB4.SA', 'B3SA3.SA', 'ABEV3.SA']
start_date = '2020-01-01'
end_date = '2024-12-31'

# === DOWNLOAD DE DADOS ===
data = yf.download(assets, start=start_date, end=end_date, interval='1D')['Close']
returns = data.pct_change().dropna()

# === 1. CLUSTERIZAÇÃO DE DIAS DE MERCADO ===
vix_sim = np.random.normal(20, 5, size=len(returns))  # VIX simulado
volatility = returns.std(axis=1)
market_return = returns.mean(axis=1)

day_features = pd.DataFrame({
    'market_return': market_return,
    'volatility': volatility,
    'vix_sim': vix_sim
}, index=returns.index)

X_day = StandardScaler().fit_transform(day_features)
kmeans_days = KMeans(n_clusters=3, random_state=42).fit(X_day)
day_features['cluster_market'] = kmeans_days.labels_

# === 2. CLUSTERIZAÇÃO DE ATIVOS ===
def compute_rsi(prices, window=14):
    delta = prices.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    return 100 - (100 / (1 + rs))

def compute_macd(prices, fast=12, slow=26, signal=9):
    ema_fast = prices.ewm(span=fast).mean()
    ema_slow = prices.ewm(span=slow).mean()
    macd = ema_fast - ema_slow
    signal_line = macd.ewm(span=signal).mean()
    return macd - signal_line

asset_features = []

for asset in assets:
    df = pd.DataFrame(index=data.index)
    df['rsi'] = compute_rsi(data[asset])
    df['macd'] = compute_macd(data[asset])
    df['return_7d'] = data[asset].pct_change(7)
    df['vol_14d'] = data[asset].pct_change().rolling(14).std()
    df['asset'] = asset
    df = df.dropna()
    asset_features.append(df)

asset_df = pd.concat(asset_features)

X_asset = StandardScaler().fit_transform(asset_df[['rsi', 'macd', 'return_7d', 'vol_14d']])
kmeans_assets = KMeans(n_clusters=4, random_state=42).fit(X_asset)
asset_df['cluster_asset'] = kmeans_assets.labels_

# === 3. MATRIZ DE ROTAÇÃO DE PORTFÓLIO ===
merged = asset_df.join(day_features['cluster_market'], how='inner')
rotation_matrix = merged.groupby(['cluster_market', 'cluster_asset'])['return_7d'].mean().unstack()

# === EXIBIR RESULTADO ===
print("\n📊 Matriz de Retornos Médios por Cluster de Mercado x Cluster de Ativo:\n")
print(rotation_matrix.round(4))
```

```python
# ✅ Análise completa do desempenho do modelo com detalhamento por split e regime
import optuna
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             brier_score_loss, confusion_matrix, classification_report)
from sklearn.cluster import KMeans
import plotly.graph_objects as go

# --- Indicadores Técnicos Sem Bibliotecas Externas ---
def rsi(prices, window=14):
    delta = prices.diff()
    up = delta.clip(lower=0)
    down = -1 * delta.clip(upper=0)
    ma_up = up.ewm(span=window).mean()
    ma_down = down.ewm(span=window).mean()
    rs = ma_up / ma_down
    return 100 - (100 / (1 + rs))

def macd_hist(prices, fast=12, slow=26, signal=9):
    ema_fast = prices.ewm(span=fast).mean()
    ema_slow = prices.ewm(span=slow).mean()
    macd = ema_fast - ema_slow
    signal_line = macd.ewm(span=signal).mean()
    return macd - signal_line

def adx(df, window=14):
    high = df['High']
    low = df['Low']
    close = df['Close']
    plus_dm = high.diff()
    minus_dm = low.diff()
    plus_dm[plus_dm < 0] = 0
    minus_dm[minus_dm > 0] = 0
    tr1 = high - low
    tr2 = abs(high - close.shift())
    tr3 = abs(low - close.shift())
    tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
    atr = tr.rolling(window=window).mean()
    plus_di = 100 * (plus_dm.rolling(window).mean() / atr)
    minus_di = abs(100 * (minus_dm.rolling(window).mean() / atr))
    dx = (abs(plus_di - minus_di) / (plus_di + minus_di)) * 100
    return dx.rolling(window).mean()

def bb_bandwidth(prices, window=20):
    sma = prices.rolling(window).mean()
    std = prices.rolling(window).std()
    upper = sma + 2 * std
    lower = sma - 2 * std
    return (upper - lower) / sma



# --- Funções para Ensemble ---
def train_ensemble_models_by_regime(train_df, features, target_col='target', regime_col='market_regime'):
    models_by_regime = {}
    for regime in sorted(train_df[regime_col].unique()):
        subset = train_df[train_df[regime_col] == regime]
        if len(subset) < 50:
            continue
        X = subset[features]
        y = subset[target_col]
        scaler = StandardScaler().fit(X)
        X_scaled = scaler.transform(X)
        models = [
            RandomForestClassifier(n_estimators=100, random_state=42),
            GradientBoostingClassifier(n_estimators=100, random_state=42),
            LogisticRegression(max_iter=1000)
        ]
        trained = [model.fit(X_scaled, y) for model in models]
        models_by_regime[regime] = {'scaler': scaler, 'models': trained}
    return models_by_regime

def predict_with_ensemble(row, models_by_regime, features, regime_col='market_regime'):
    regime = row[regime_col]
    if regime not in models_by_regime:
        return 0.5
    info = models_by_regime[regime]

    # 👉 Criar DataFrame com nomes corretos
    X = pd.DataFrame([row[features].values], columns=features)

    X_scaled = info['scaler'].transform(X)
    probs = [model.predict_proba(X_scaled)[0][1] for model in info['models']]
    return np.mean(probs)


# --- Backtest com avaliação detalhada ---
def evaluate_predictions(y_true, y_pred, y_prob):
    return {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, zero_division=0),
        'recall': recall_score(y_true, y_pred, zero_division=0),
        'f1': f1_score(y_true, y_pred, zero_division=0),
        'brier': brier_score_loss(y_true, y_prob)
    }

def backtest_ensemble_analysis(df, features, regime_col='market_regime', target_col='target', price_col='Close', n_splits=5):
    total_size = len(df)
    split_size = total_size // (n_splits + 1)
    all_metrics = []
    all_curves = []

    for i in range(n_splits):
        train_end = (i + 1) * split_size
        test_start = train_end
        test_end = train_end + split_size
        train = df.iloc[:train_end]
        test = df.iloc[test_start:test_end].copy()

        models_by_regime = train_ensemble_models_by_regime(train, features, target_col, regime_col)
        test['prob'] = test.apply(lambda row: predict_with_ensemble(row, models_by_regime, features, regime_col), axis=1)

        # prob_buy          0.520000
        # prob_sell         0.480000
        # adx_min          15.000000
        # rsi_max_buy      65.000000
        # rsi_min_sell     45.000000
        # macd_min_buy     -0.100000
        # macd_max_sell     0.100000
        # avg_sharpe        0.800367
        def decide(row):
            if row['prob'] > 0.52  and row['rsi'] < 65 and row['macd_hist'] > -0.10 and row['adx'] > 15:
                return 1
            elif row['prob'] < 0.48  and row['rsi'] > 45 and row['macd_hist'] < 0.10 and row['adx'] > 15:
                return 0
            else:
                return np.nan

        test['signal'] = test.apply(decide, axis=1)
        test['signal'].fillna(method='ffill', inplace=True)
        test['returns'] = test[price_col].pct_change().fillna(0)
        test['strategy_returns'] = test['returns'] * test['signal'].shift(1).fillna(0)
        test['cumulative'] = (1 + test['strategy_returns']).cumprod()
        test['cummax'] = test['cumulative'].cummax()
        test['drawdown'] = (test['cumulative'] - test['cummax']) / test['cummax']

        y_true = test[target_col]
        y_pred = (test['prob'] > 0.5).astype(int)
        metrics = evaluate_predictions(y_true, y_pred, test['prob'])
        sharpe = (test['strategy_returns'].mean() / test['strategy_returns'].std()) * np.sqrt(252) if test['strategy_returns'].std() > 0 else 0

        metrics.update({
            'split': i + 1,
            'retorno_%': test['cumulative'].iloc[-1] * 100 - 100,
            'drawdown_%': test['drawdown'].min() * 100,
            'sharpe': sharpe,
            'trades': test['signal'].nunique()
        })
        all_metrics.append(metrics)
        all_curves.append(test[['cumulative']].rename(columns={'cumulative': f'Split {i+1}'}))

    fig = go.Figure()
    for curve in all_curves:
        name = curve.columns[0]
        fig.add_trace(go.Scatter(x=curve.index, y=curve[name], mode='lines', name=name))

    fig.update_layout(title='📈 Curva de Capital por Split', xaxis_title='Data', yaxis_title='Capital Acumulado',
                      template='plotly_dark', height=500)
    fig.show()
    df_all_metrics = pd.DataFrame(all_metrics).to_csv('all_metrics.csv')
    return pd.DataFrame(all_metrics)

```

```python
import pandas as pd
import numpy as np
import yfinance as yf
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")

# 1. Baixar dados
def load_data(ticker="^HSI", period="10y"):
    df = yf.download(ticker, period=period, interval="1d")
    df = df.xs((ticker),level="Ticker",axis=1)
    return df.dropna()

# 2. Indicadores técnicos
def calculate_indicators(df):
    df['returns_pct'] = df['Close'].pct_change()
    df['volatility'] = df['returns_pct'].rolling(10).std()
    df['volatility_20d'] = df['returns_pct'].rolling(20).std()
    df['momentum'] = df['Close'].pct_change(10)
    df['rolling_sharpe'] = df['returns_pct'].rolling(20).mean() / df['returns_pct'].rolling(20).std()
    df['drawdown'] = (df['Close'] / df['Close'].cummax()) - 1

    # RSI
    delta = df['Close'].diff()
    up = delta.clip(lower=0)
    down = -delta.clip(upper=0)
    ma_up = up.rolling(14).mean()
    ma_down = down.rolling(14).mean()
    rs = ma_up / ma_down
    df['rsi'] = 100 - (100 / (1 + rs))

    # MACD
    ema12 = df['Close'].ewm(span=12).mean()
    ema26 = df['Close'].ewm(span=26).mean()
    df['macd'] = ema12 - ema26
    df['macd_signal'] = df['macd'].ewm(span=9).mean()

    # Bollinger
    boll_mid = df['Close'].rolling(20).mean()
    boll_std = df['Close'].rolling(20).std()
    df['bollinger_width'] = 2 * boll_std / boll_mid

    df['year'] = df.index.year
    df['month'] = df.index.month
    df['dayofweek'] = df.index.dayofweek

    # Identificar resistência Alta
    df['resistance_high'] = df['High'].rolling(window=50).max()

    # Identificar resistência Baixa
    df['resistance_low'] = df['Low'].rolling(window=50).min()


    df.dropna(inplace=True)
    return df

# 3. Target
def add_target(df):
    df['return_3d_pct'] = df['Close'].shift(-3) / df['Close'] - 1
    df['target'] = df['return_3d_pct'].apply(lambda x: 1 if x > 0 else 0)
    df.dropna(inplace=True)
    return df

# 4. Modelos por era
def train_models_by_eras(df, features, eras):
    models = {}
    scaler = StandardScaler()  # Adicionado: escalonador para features
    for era_range in eras:
        df_era = df[df.index.to_series().between(*era_range)]
        X = df_era[features].replace([np.inf, -np.inf], 0).fillna(0)
        y = df_era['target']
        # Escalonamento das features
        X_scaled = scaler.fit_transform(X)
        model = XGBClassifier(use_label_encoder=False, eval_metric="logloss")
        model.fit(X_scaled, y)
        models[era_range] = (model, scaler)  # Armazenar modelo e scaler
    return models


# 5. Geração de previsões de todos os modelos
def generate_model_predictions(df, models_by_era, features):
    df = df.copy()
    for era_range, (model, scaler) in models_by_era.items():
        name = f"pred_{era_range[0][:4]}"
        X = df[features].replace([np.inf, -np.inf], 0).fillna(0)
        X_scaled = scaler.transform(X)  # Aplicar escalonamento
        df[name] = model.predict(X_scaled)
    return df


# 6. Meta-features
def generate_meta_features(df):
    regime_data = df[['returns_pct', 'volatility', 'momentum', 'drawdown']].dropna()
    # Adicionado: normalização antes do KMeans
    scaler = StandardScaler()
    regime_data_scaled = scaler.fit_transform(regime_data)
    kmeans = KMeans(n_clusters=3, random_state=42).fit(regime_data_scaled)
    df['regime_cluster'] = kmeans.predict(scaler.transform(df[['returns_pct', 'volatility', 'momentum', 'drawdown']].fillna(0)))
    return df

# 7. Treinar meta-modelo
def train_meta_model(df, meta_features, label_col='target'):
    df = df.dropna(subset=meta_features + [label_col])
    X = df[meta_features + [label_col]]
    y = df[label_col]

    # Divisão temporal fixa
    start_train = "2015-01-01"
    end_train = "2018-12-31"
    start_test = "2019-01-01"
    end_test = "2024-12-01"
    df1_train = X.loc[start_train:end_train]
    df1_test = X.loc[start_test:end_test]
    X_train = df1_train[meta_features]
    y_train = df1_train[label_col]
    X_test = df1_test[meta_features]
    y_test = df1_test[label_col]

    # Escalonamento das meta-features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    meta_model = LGBMClassifier(num_leaves=63,random_state=42)
    meta_model.fit(X_train_scaled, y_train)
    preds = meta_model.predict(X_test_scaled)
    acc = accuracy_score(y_test, preds)
    print(f"🎯 Meta-modelo Acurácia: {acc:.4f}")
    return meta_model, X_test.index, preds, scaler

# 8. Aplicar e avaliar
def apply_and_evaluate(df, meta_model, meta_features, pred_indexes, scaler):
    df_result = df.loc[pred_indexes].copy()
    X = df_result[meta_features]
    X_scaled = scaler.transform(X)  # Aplicar escalonamento
    df_result['meta_pred'] = meta_model.predict(X_scaled)

    # Adicionado: incorporar custo de transação
    transaction_cost = 0.001  # 0.1% por transação
    df_result['strategy_return'] = np.where(df_result['meta_pred'] == 1,
                                           df_result['return_3d_pct'] - transaction_cost,
                                           -df_result['return_3d_pct'] - transaction_cost)
    df_result['buyhold_return'] = df_result['return_3d_pct']
    df_result['meta_cum'] = (1 + df_result['strategy_return']).cumprod()
    df_result['buyhold_cum'] = (1 + df_result['buyhold_return']).cumprod()

    # Adicionado: cálculo de métricas adicionais
    annual_ret = df_result['strategy_return'].mean() * 252  # 252 dias úteis
    annual_std = df_result['strategy_return'].std() * np.sqrt(252)
    sharpe = annual_ret / annual_std if annual_std != 0 else 0
    max_drawdown = df_result['meta_cum'].div(df_result['meta_cum'].cummax()).sub(1).min()
    print(f"📊 Sharpe Ratio: {sharpe:.4f}, Max Drawdown: {max_drawdown:.4f}")

    return df_result

# Pipeline completo
df = load_data('BOVA11.SA')
df = calculate_indicators(df)
df = add_target(df)

features = ['rsi',  'bollinger_width', 'volatility_20d',
            'year', 'month',  'momentum', 'rolling_sharpe', 'drawdown', 'resistance_high',
            'resistance_low']

eras = {
    ("2013-01-01", "2016-12-31"),
    ("2017-01-01", "2019-12-31"),
    ("2020-01-01", "2025-01-01")
}

models = train_models_by_eras(df, features, eras)
df = generate_model_predictions(df, models, features)
df = generate_meta_features(df)

# Meta-features = previsões dos modelos + contexto atual
meta_features = [f"pred_{era[0][:4]}" for era in eras] + ['volatility', 'momentum', 'drawdown', 'regime_cluster']
meta_model, pred_highlights, preds, scaler = train_meta_model(df, meta_features)
df_result = apply_and_evaluate(df, meta_model, meta_features, pred_highlights, scaler)
```

```python
def calculate_rsi(data, periods=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=periods).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=periods).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def find_resistance(data, window=50):
    return data['High'].rolling(window=window).max()
```

```python
def walk_forward_best_ensemble_with_profit(X, y, df, window_size=60, test_size=10, n_trials=10):
    results = []
    conf_matrix_sum = np.zeros((2,2))
    profits = []

    for start in range(0, len(X) - window_size - test_size + 1, test_size):
        X_train = X[start:start + window_size]
        y_train = y[start:start + window_size]
        X_test = X[start + window_size:start + window_size + test_size]
        y_test = y[start + window_size:start + window_size + test_size]

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        models = {}

        # SVM
        def objective_svm(trial):
            C = trial.suggest_loguniform('C', 1e-3, 1e3)
            gamma = trial.suggest_loguniform('gamma', 1e-4, 1e1)
            model = SVC(C=C, gamma=gamma, kernel='rbf', probability=True)
            model.fit(X_train_scaled, y_train)
            return accuracy_score(y_train, model.predict(X_train_scaled))

        study_svm = optuna.create_study(direction='maximize')
        study_svm.optimize(objective_svm, n_trials=n_trials, show_progress_bar=False)
        svm = SVC(**study_svm.best_params, kernel='rbf', probability=True)
        svm.fit(X_train_scaled, y_train)
        models['SVM'] = (svm, svm.predict_proba(X_test_scaled), accuracy_score(y_train, svm.predict(X_train_scaled)))

        # KNN
        def objective_knn(trial):
            n = trial.suggest_int('n_neighbors', 1, 20)
            model = KNeighborsClassifier(n_neighbors=n)
            model.fit(X_train_scaled, y_train)
            return accuracy_score(y_train, model.predict(X_train_scaled))

        study_knn = optuna.create_study(direction='maximize')
        study_knn.optimize(objective_knn, n_trials=n_trials, show_progress_bar=False)
        knn = KNeighborsClassifier(**study_knn.best_params)
        knn.fit(X_train_scaled, y_train)
        models['KNN'] = (knn, knn.predict_proba(X_test_scaled), accuracy_score(y_train, knn.predict(X_train_scaled)))

        # Tree
        def objective_tree(trial):
            d = trial.suggest_int('max_depth', 2, 10)
            model = DecisionTreeClassifier(max_depth=d)
            model.fit(X_train_scaled, y_train)
            return accuracy_score(y_train, model.predict(X_train_scaled))

        study_tree = optuna.create_study(direction='maximize')
        study_tree.optimize(objective_tree, n_trials=n_trials, show_progress_bar=False)
        tree = DecisionTreeClassifier(**study_tree.best_params)
        tree.fit(X_train_scaled, y_train)
        models['Tree'] = (tree, tree.predict_proba(X_test_scaled), accuracy_score(y_train, tree.predict(X_train_scaled)))

        # RF
        def objective_rf(trial):
            n = trial.suggest_int('n_estimators', 10, 100)
            d = trial.suggest_int('max_depth', 2, 10)
            model = RandomForestClassifier(n_estimators=n, max_depth=d)
            model.fit(X_train_scaled, y_train)
            return accuracy_score(y_train, model.predict(X_train_scaled))

        study_rf = optuna.create_study(direction='maximize')
        study_rf.optimize(objective_rf, n_trials=n_trials, show_progress_bar=False)
        rf = RandomForestClassifier(**study_rf.best_params)
        rf.fit(X_train_scaled, y_train)
        models['RF'] = (rf, rf.predict_proba(X_test_scaled), accuracy_score(y_train, rf.predict(X_train_scaled)))

        # LR
        def objective_lr(trial):
            c = trial.suggest_loguniform('C', 1e-3, 1e3)
            model = LogisticRegression(C=c, solver='liblinear')
            model.fit(X_train_scaled, y_train)
            return accuracy_score(y_train, model.predict(X_train_scaled))

        study_lr = optuna.create_study(direction='maximize')
        study_lr.optimize(objective_lr, n_trials=n_trials, show_progress_bar=False)
        lr = LogisticRegression(**study_lr.best_params, solver='liblinear')
        lr.fit(X_train_scaled, y_train)
        models['LR'] = (lr, lr.predict_proba(X_test_scaled), accuracy_score(y_train, lr.predict(X_train_scaled)))

        # XGB
        xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
        xgb.fit(X_train_scaled, y_train)
        models['XGB'] = (xgb, xgb.predict_proba(X_test_scaled), accuracy_score(y_train, xgb.predict(X_train_scaled)))

        # LGBM
        lgbm = LGBMClassifier()
        lgbm.fit(X_train_scaled, y_train)
        models['LGBM'] = (lgbm, lgbm.predict_proba(X_test_scaled), accuracy_score(y_train, lgbm.predict(X_train_scaled)))

        # Testar combinações
        best_acc = 0
        best_combo = None
        best_preds = None

        keys = list(models.keys())
        for r_size in range(2, len(keys)+1):
            for combo in combinations(keys, r_size):
                accs = np.array([models[m][2] for m in combo])
                acc_total = accs.sum()
                if acc_total == 0:
                    continue

                probs_weighted = sum(models[m][1] * (accs[i]/acc_total) for i, m in enumerate(combo))
                preds = np.argmax(probs_weighted, axis=1)
                acc = accuracy_score(y_test, preds)

                if acc > best_acc:
                    best_acc = acc
                    best_combo = combo
                    best_preds = preds

        if best_preds is None:
            best_model_key = max(models, key=lambda m: models[m][2])
            best_preds = np.argmax(models[best_model_key][1], axis=1)
            best_combo = (best_model_key,)
            best_acc = accuracy_score(y_test, best_preds)

        # Calcular lucro
        # Calcular lucro (somente long)
        idx_test_start = start + window_size
        for j, pred in enumerate(best_preds):
            idx = idx_test_start + j
            if idx + 1 >= len(df):
                continue
            if pred == 1:
                open_price = df['Open'].iloc[idx + 1]
                close_price = df['Close'].iloc[idx + 1]
                profit = close_price - open_price
                profits.append(profit)

        cm = confusion_matrix(y_test, best_preds, labels=[0,1])
        conf_matrix_sum += cm

        results.append({
            'start': start,
            'acc_test': best_acc,
            'f1_test': f1_score(y_test, best_preds, zero_division=0),
            'best_combo': best_combo
        })

        print(f"Janela {start}: acc = {best_acc:.2f}, best combo = {best_combo}")

    conf_matrix_avg = conf_matrix_sum / len(results)
    return results, conf_matrix_avg, profits
```

```python
#!/usr/bin/env python
# coding: utf-8

import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from collections import Counter
import matplotlib.pyplot as plt
import datetime
from sklearn.metrics import accuracy_score

# === Funções de Indicadores Técnicos ===
def calculate_rsi(data, periods=14):
    """Calcula o Índice de Força Relativa (RSI)."""
    delta = data.diff()
    gain = delta.where(delta > 0, 0).rolling(window=periods).mean()
    loss = -delta.where(delta < 0, 0).rolling(window=periods).mean()
    rs = gain / loss
    rs = rs.replace([np.inf, -np.inf], np.nan)
    return 100 - (100 / (1 + rs))

def add_technical_indicators(df):
    """Adiciona indicadores técnicos ao DataFrame, incluindo os da saída fornecida."""
    df = df.copy()

    # Indicadores padrão
    df['volume_change'] = df['Volume'].pct_change()
    df['ma20'] = df['Close'].rolling(20).mean()
    df['std20'] = df['Close'].rolling(20).std()
    df['upper_band'] = df['ma20'] + 2 * df['std20']
    df['lower_band'] = df['ma20'] - 2 * df['std20']
    bb_range = df['upper_band'] - df['lower_band']
    bb_range = bb_range.replace(0, np.nan)
    df['percent_b'] = (df['Close'] - df['lower_band']) / bb_range
    df['momentum'] = df['Close'] - df['Close'].shift(10)
    df['rsi'] = calculate_rsi(df['Close'])

    # Volatilidade histórica (20 dias)
    df['hist_volatility'] = df['Close'].pct_change().rolling(20).std() * np.sqrt(252)

    # RSI ajustado por ATR
    df['tr'] = np.maximum(df['High'] - df['Low'],
                         np.maximum(abs(df['High'] - df['Close'].shift()),
                                   abs(df['Low'] - df['Close'].shift())))
    df['atr'] = df['tr'].rolling(14).mean()
    df['rsi_atr'] = df['rsi'] / df['atr']

    # Volume relativo
    df['volume_relative'] = df['Volume'] / df['Volume'].rolling(20).mean()

    # Posição de pivô
    df['pivot'] = (df['High'].rolling(20).max() + df['Low'].rolling(20).min() + df['Close'].rolling(20).mean()) / 3
    df['pivot_position'] = (df['Close'] - df['pivot']) / df['pivot']

    # ADX
    high = df['High']
    low = df['Low']
    close = df['Close']
    df['plus_dm'] = np.where((high - high.shift()) > (low.shift() - low), np.maximum(high - high.shift(), 0), 0)
    df['minus_dm'] = np.where((low.shift() - low) > (high - high.shift()), np.maximum(low.shift() - low, 0), 0)
    plus_di = 100 * (df['plus_dm'].rolling(14).sum() / df['atr'])
    minus_di = 100 * (df['minus_dm'].rolling(14).sum() / df['atr'])
    df['adx'] = 100 * abs(plus_di - minus_di) / (plus_di + minus_di)

    # Prop (proporção de dias de alta em 20 dias)
    df['Prop'] = df['Close'].pct_change().apply(lambda x: 1 if x > 0 else 0).rolling(20).mean()

    return df

# === Carregamento de Dados ===
ticker = "^HSI"
end_date = datetime.datetime(2025, 7, 22)  # Data atual
start_date = datetime.datetime(2018, 1, 1)
data = yf.download(ticker, start=start_date, end=end_date)
df = data.xs((ticker),level="Ticker",axis=1)
df = df[['Open', 'High', 'Low', 'Close', 'Volume']].copy()

# Adiciona indicadores técnicos
df = add_technical_indicators(df)
df['Regime'] = np.where(df['adx'] > 20, 1, 0)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# === Definição do Target ===
future = 5
df['target_5d'] = df['Close'].shift(-future) - df['Close']
epsilon = 0.01  # Tolerância de 1% do preço atual
df['target'] = df.apply(lambda row: 1 if row['target_5d'] > epsilon * row['Close'] else -1 if row['target_5d'] < -epsilon * row['Close'] else 0, axis=1)
# Ajustar target para XGBoost (0, 1, 2 em vez de -1, 0, 1)
df['target'] = df['target'].map({-1: 0, 0: 1, 1: 2})

# === Seleção de Features ===
features = ['hist_volatility', 'ma20', 'std20', 'rsi_atr', 'percent_b']  # Top 5 features
X_all = df[features]
y_all = df['target']
dates = df.index

# === Walk-forward + Verificação de Overfitting ===
train_window = 600
test_window = future
step_size = future
search_log = []
train_accuracies = []
test_accuracies = []
feature_importances = []

y_all.value_counts()
```

```python
for i in range(train_window, len(df) - test_window, step_size):
    X_train = X_all.iloc[i - train_window:i]
    y_train = y_all.iloc[i - train_window:i]
    X_test = X_all.iloc[i:i + test_window]
    y_test = y_all.iloc[i:i + test_window]

    # Garantir que todas as classes (0, 1, 2) estejam presentes
    if len(y_train.value_counts()) != 3 or y_train.value_counts().min() < 10:
        continue
    # Normalização
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Calcular peso para classes desbalanceadas
    class_counts = y_train.value_counts()
    n_samples = len(y_train)
    n_classes = len(class_counts)
    scale_pos_weight = {i: n_samples / (n_classes * class_counts.get(i, 1)) for i in range(3)}

    # RandomizedSearchCV com XGBoost
    param_dist = {
        'n_estimators': [50, 100],
        'max_depth': [3, 4],
        'min_child_weight': [3, 5, 7],
        'gamma': [0.1, 0.3],
        'reg_lambda': [1, 2],
        'learning_rate': [0.01, 0.1]
    }
    search = RandomizedSearchCV(
        estimator=XGBClassifier(random_state=42, objective='multi:softmax', num_class=3),
        param_distributions=param_dist,
        n_iter=15,
        scoring='accuracy',
        cv=TimeSeriesSplit(n_splits=7),
        n_jobs=-1,
        random_state=42
    )
    search.fit(X_train_scaled, y_train, sample_weight=[scale_pos_weight[y] for y in y_train])
    search_log.append({
        'index': i,
        'date': df.index[i],
        'best_params': search.best_params_,
        'best_score': search.best_score_
    })

    # Avaliar acurácia de treino e teste
    best_model = search.best_estimator_
    train_pred = best_model.predict(X_train_scaled)
    test_pred = best_model.predict(X_test_scaled)
    train_accuracy = accuracy_score(y_train, train_pred)
    test_accuracy = accuracy_score(y_test, test_pred)
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

    # Importância das features
    feature_importances.append(best_model.feature_importances_)

```

```python
# === Análise de Overfitting ===
avg_train_accuracy = np.mean(train_accuracies)
avg_test_accuracy = np.mean(test_accuracies)
accuracy_diff = avg_train_accuracy - avg_test_accuracy

print("\n=== Análise de Overfitting ===")
print(f"Acurácia média no treino: {avg_train_accuracy:.2%}")
print(f"Acurácia média no teste: {avg_test_accuracy:.2%}")
print(f"Diferença (treino - teste): {accuracy_diff:.2%}")
if accuracy_diff > 0.1:
    print("⚠️ Possível overfitting: A acurácia no treino é significativamente maior que no teste.")
else:
    print("✅ Sem sinais claros de overfitting: A diferença entre treino e teste é aceitável.")

# === Estabilidade dos Hiperparâmetros ===
log_df = pd.DataFrame(search_log)
param_counts = Counter([str(params) for params in log_df['best_params']]).most_common()
print("\nEstabilidade dos hiperparâmetros:")
for params, count in param_counts[:3]:
    print(f"Parâmetros: {params}, Frequência: {count} ({count/len(log_df):.2%})")

# === Importância Média das Features ===
avg_feature_importance = np.mean(feature_importances, axis=0)
feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': avg_feature_importance})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print("\nImportância média das features:")
print(feature_importance_df)
```

```python
# === Backtest com Modelo Fixo ===
top_params = eval(param_counts[0][0])  # Converte string para dicionário
preds, confs, pos, fut_ret = [], [], [], []
for i in range(train_window, len(df) - test_window, step_size):
    X_train = X_all.iloc[i - train_window:i]
    y_train = y_all.iloc[i - train_window:i]
    X_test = X_all.iloc[i:i + test_window]

    if y_train.value_counts().min() < 10:
        continue

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    class_counts = y_train.value_counts()
    n_samples = len(y_train)
    n_classes = len(class_counts)
    scale_pos_weight = {i: n_samples / (n_classes * class_counts.get(i, 1)) for i in range(3)}

    model = XGBClassifier(**top_params, random_state=42, objective='multi:softmax', num_class=3)
    model.fit(X_train_scaled, y_train, sample_weight=[scale_pos_weight[y] for y in y_train])
    proba = model.predict_proba(X_test_scaled)
    pred = model.predict(X_test_scaled)
    conf = proba.max(axis=1)
    pred_filtered = [p if c >= 0.6 else 1 for p, c in zip(pred, conf)]  # 1 é 'Neutro'

    final_signals = []
    last_pos = 1  # Inicia como neutro
    for sig, c in zip(pred_filtered, conf):
        if sig != last_pos and c >= 0.65:
            last_pos = sig
        final_signals.append(last_pos)

    position = [1] + final_signals[:-1]
    future_returns = df['Close'].pct_change().shift(-1).iloc[i:i + test_window]
    # Ajustar sinais para calcular retorno (-1, 0, 1)
    position_adjusted = [1 if p == 2 else -1 if p == 0 else 0 for p in position]
    preds.extend(pred_filtered)
    confs.extend(conf)
    pos.extend(position_adjusted)
    fut_ret.extend(future_returns)

# === Avaliação do Backtest ===
walk_df = df.iloc[train_window:train_window + len(preds)].copy()
walk_df['pred'] = [0 if p == 0 else 1 if p == 2 else p for p in preds]  # Ajustar predições
walk_df['confidence'] = confs
walk_df['position'] = pos
walk_df['future_return'] = fut_ret
walk_df.dropna(inplace=True)

walk_df['strategy_return'] = walk_df['position'] * walk_df['future_return']
walk_df['cumulative_strategy'] = (1 + walk_df['strategy_return']).cumprod()
walk_df['cumulative_buyhold'] = (1 + walk_df['future_return']).cumprod()

total_strategy = walk_df['cumulative_strategy'].iloc[-1] - 1
total_buyhold = walk_df['cumulative_buyhold'].iloc[-1] - 1
std_dev = walk_df['strategy_return'].std()
sharpe = walk_df['strategy_return'].mean() / std_dev * np.sqrt(252) if std_dev != 0 else np.nan
vol = std_dev * np.sqrt(252)
dd = (walk_df['cumulative_strategy'] / walk_df['cumulative_strategy'].cummax() - 1).min()

print("\n=== Resultados do Backtest ===")
print(f"Retorno total estratégia: {total_strategy:.2%}")
print(f"Retorno total Buy & Hold: {total_buyhold:.2%}")
print(f"Sharpe Ratio: {sharpe:.2f}")
print(f"Máx Drawdown: {dd:.2%}")
print(f"Volatilidade anualizada: {vol:.2%}")

# === Gráfico do Backtest ===
plt.figure(figsize=(12, 6))
plt.plot(walk_df.index, walk_df['cumulative_strategy'], label='Modelo (XGBoost)')
plt.plot(walk_df.index, walk_df['cumulative_buyhold'], label='Buy & Hold')
plt.title('Backtest com Sinal Estabilizado (XGBoost)')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

```

```python
# RSI (14 períodos)
def calculate_rsi(data, periods=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=periods).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=periods).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# Identificar resistência
def find_resistance(data, window=50):
    resistance = data['High'].rolling(window=window).max()
    return resistance

# Identificar resistência
def find_support(data, window=50):
    resistance = data['Low'].rolling(window=window).min()
    return resistance

def drop_highly_correlated_features(df, threshold=0.7):
    # Apenas colunas numéricas
    corr_matrix = df.corr(numeric_only=True).abs()

    # Triângulo superior da matriz de correlação
    upper = corr_matrix.where(
        ~pd.DataFrame(np.tril(np.ones(corr_matrix.shape)).astype(bool), columns=corr_matrix.columns, index=corr_matrix.index)
    )

    # Achar colunas com correlação acima do limiar
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

    print(f"📉 Removendo {len(to_drop)} features com correlação > {threshold}: {to_drop}")
    return df.drop(columns=to_drop)

def topos_fundos(data):
    # Inicialize listas vazias para armazenar os topos e fundos
    topos = []
    fundos = []
    # Percorra a série de preços para identificar topos e fundos
    for i in range(1, len(data) - 1):
        if data["Close"][i] > data["Close"][i - 1] and data["Close"][i] > data["Close"][i + 1]:
            topos.append((data.index[i], data["Close"][i]))
        elif data["Close"][i] < data["Close"][i - 1] and data["Close"][i] < data["Close"][i + 1]:
            fundos.append((data.index[i], data["Close"][i]))
    data["Topo"] = False
    data["Fundo"] = False
    for topo in topos:
        data.loc[topo[0], "Topo"] = True

    for fundo in fundos:
        data.loc[fundo[0], "Fundo"] = True
    # Converter True e False em 1 e 0
    data["Topo"] = data["Topo"].astype(int)
    data["Fundo"] = data["Fundo"].astype(int)


# RSI
def rsi(series, period=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def atr(df):
     # ADX simplificado
    high = df['High']
    low = df['Low']
    close = df['Close']
    df['tr'] = np.maximum(high - low, np.maximum(abs(high - close.shift()), abs(low - close.shift())))
    df['atr'] = df['tr'].rolling(14).mean()
    df['plus_dm'] = np.where((high - high.shift()) > (low.shift() - low), np.maximum(high - high.shift(), 0), 0)
    df['minus_dm'] = np.where((low.shift() - low) > (high - high.shift()), np.maximum(low.shift() - low, 0), 0)
    plus_di = 100 * (df['plus_dm'].rolling(14).sum() / df['atr'])
    minus_di = 100 * (df['minus_dm'].rolling(14).sum() / df['atr'])
    df['adx'] = 100 * abs(plus_di - minus_di) / (plus_di + minus_di)

# === 2. Indicadores técnicos robustos ===
def add_technical_indicators(df, price_col='Close'):
    df = df.copy()

    df['return'] = df['Close'].pct_change()
    df['volume_change'] = df['Volume'].pct_change()

    # 1. Médias móveis
    df['ma_5'] = df[price_col].rolling(5).mean()
    df['ma_21'] = df[price_col].rolling(21).mean()
    df['ma_diff'] = df['ma_5'] - df['ma_21']

    # 2. Slope da tendência (reta de regressão)
    df['slope_5'] = df[price_col].rolling(5).apply(
        lambda x: np.polyfit(range(len(x)), x, 1)[0], raw=False)

    df['slope_21'] = df[price_col].rolling(21).apply(
        lambda x: np.polyfit(range(len(x)), x, 1)[0], raw=False)

    df['slope_50'] = df[price_col].rolling(50).apply(
        lambda x: np.polyfit(range(len(x)), x, 1)[0], raw=False)

    # 3. Retorno acumulado (momentum)
    df['ret_5'] = df[price_col].pct_change(5)
    df['ret_10'] = df[price_col].pct_change(10)
    df['ret_21'] = df[price_col].pct_change(21)

    # 4. Candles consecutivos na mesma direção
    df['direction'] = np.sign(df[price_col].diff())
    df['same_dir_count'] = (
        df['direction']
        .groupby((df['direction'] != df['direction'].shift()).cumsum())
        .cumcount() + 1
    )

    # 5. Z-score (desvio do preço da média)
    df['zscore_21'] = (df[price_col] - df['ma_21']) / df[price_col].rolling(21).std()

   # 6. Ângulo da linha de tendência
    slopes = df[price_col].rolling(10).apply(
        lambda x: np.polyfit(range(len(x)), x, 1)[0], raw=False)
    df['trend_angle_deg'] = np.degrees(np.arctan(slopes))

    # Prop
    df["Prop"] = (df["Close"]-df["Open"])/(df["High"]-df["Low"])

    topos_fundos(df)

    df['MM15'] = df["Close"].rolling(15).mean()
    df['No_Trend'] = df["Close"] - df["MM15"]
    df['LAG1'] =   df["No_Trend"].shift(1)
    df['LAG2'] =   df["No_Trend"].shift(2)
    df['LAG3'] =   df["No_Trend"].shift(3)
    df['Volume_mean'] = df['Volume']/df["Volume"].rolling(15).mean()  -1

    price_range = df['High'] - df['Low']
    price_range.rolling(window=15).mean()

    atr(df)

    df['PMA'] = price_range

    df['rsi'] = rsi(df['Close'])

     # Preço normalizado
    range_ = df['Close'].rolling(20).max() - df['Close'].rolling(20).min()
    range_ = range_.replace(0, np.nan)
    df['price_pos'] = (df['Close'] - df['Close'].rolling(20).min()) / range_

     # Suporte e resitencia
    df['resistencia'] = find_resistance(df)
    df['support'] = find_support(df)

    # Candle de alta
    df['candle_alta'] = (df['Close'] - df['Open']) / (df['High'] - df['Low']) > 0.75

    # Adicionar volatilidade histórica (proxy para volatilidade implícita)
    df['hist_volatility'] = df['return'].rolling(window=20).std() * np.sqrt(252)  # Volatilidade anualizada

    # RSI Normalizado por Volatilidade (RSI/ATR)
    df['rsi_atr'] = df['rsi'] / df['atr']


    # Drop intermediários
    df.drop(columns=['direction'], inplace=True)


    return df

```

```python
import pandas as pd
import plotly.express as px

# === 1. Função para formatar SHAP values por regime + target ===
def plot_shap_plotly(df_shap_results, regime=0, target_day="direction_D1", top_n=10):
    # Filtrar pelos parâmetros
    df_filtered = df_shap_results[
        (df_shap_results['regime'] == regime) &
        (df_shap_results['target_day'] == target_day)
    ]

    # Calcula média dos SHAP values absolutos por feature
    shap_mean = df_filtered.drop(columns=['true', 'pred', 'target_day', 'regime']).abs().mean().sort_values(ascending=False)
    top_features = shap_mean.head(top_n).index.tolist()

    # Cria dataframe com valores médios reais (não absolutos)
    shap_df = df_filtered[top_features].mean().reset_index()
    shap_df.columns = ['feature', 'impact']
    shap_df = shap_df.sort_values(by='impact', key=abs, ascending=True)

    # Plotly horizontal bar
    fig = px.bar(
        shap_df,
        x='impact',
        y='feature',
        orientation='h',
        color='impact',
        color_continuous_scale='RdBu',
        title=f"SHAP Impacto Médio - Regime {regime} - {target_day}",
        labels={'impact': 'SHAP Value'}
    )

    fig.update_layout(yaxis_title='Feature', xaxis_title='Impacto Médio SHAP', height=500)
    fig.show()

# === 2. Exemplo de uso ===
plot_shap_plotly(df_shap_results, regime=1, target_day="direction_D3", top_n=10)
```
